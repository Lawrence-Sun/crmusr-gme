现在要处理的事情就是把之前的扣点数据生成gps能够直接读取的文件的程序转写到这里

```python
def convert_to_gsf_config(original_data_name, output_data_name, mode):
    import pandas as pd

    original_data = pd.read_csv(original_data_name, sep=",", header=None)
    sorted_data = original_data.sort_values(by=0)
    # print(sorted_data)

    gps_column = []

    for i in range(0, len(sorted_data)):
        gps_column.append("/gps/hist/point")
        MeV_energy = sorted_data.iloc[i, 0] * 1000  # Convert from MeV to GeV
        if mode == "flux":
            flux = sorted_data.iloc[i, 1]
        elif mode == "2.7":
            flux = sorted_data.iloc[i, 1] / (sorted_data.iloc[i, 0]**2.7)
            # flux = sorted_data.iloc[i, 1] / (MeV_energy**2.7)
            # flux = sorted_data.iloc[i, 1] / (MeV_energy)
        elif mode == "2.6":
            flux = sorted_data.iloc[i, 1] / (sorted_data.iloc[i, 0]**2.6)

        sorted_data.iloc[i, 0] = MeV_energy
        sorted_data.iloc[i, 1] = flux

    # sorted_data.insert(0, "header", gps_column)
    sorted_data.drop(sorted_data.columns[1], axis=1, inplace=True)
    # print(sorted_data)

    sorted_data.to_csv(output_data_name, sep=" ", header=False, index=False)
```

```python
convert_to_gsf_config("current_version/25_GSF_proton.csv", "current_version/25_GSF_proton_gps.csv", "2.6")
convert_to_gsf_config("current_version/25_GSF_alpha.csv", "current_version/25_GSF_alpha_gps.csv", "2.6")
```
